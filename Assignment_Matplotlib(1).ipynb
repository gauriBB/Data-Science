#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in a wide range of formats. It provides a convenient interface for generating various types of plots and charts, making it an essential tool for data visualization and scientific plotting in Python.
 
Matplotlib is primarily used to visualize data in the form of graphs, charts, and plots, making it easier to interpret and analyze the underlying patterns and trends.

Customization: It offers extensive customization options to fine-tune the appearance of plots, such as adjusting colors, styles, markers, labels, and more.

Support for different plot types: Matplotlib supports a wide variety of plot types, ranging from basic line plots to complex 3D visualizations, allowing users to handle diverse data scenarios.

Line plot, Scatter plot, Bar plot, Histogram, Pie chart are the  plots that can be plotted using the Pyplot module of Matplotlib.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Plot')
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3, 4, 5])
y1= np.array([0, 100, 200, 300, 400, 500]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
# Create a 2x2 grid for subplots
plt.subplot(2, 2, 1)  # Top-left subplot
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.subplot(2, 2, 2)  # Top-right subplot
plt.plot(x, y2, label='Line 2', color='orange')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2, 2, 3)  # Bottom-left subplot
plt.plot(x, y3, label='Line 3', color='green')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2, 2, 4)  # Bottom-right subplot
plt.plot(x, y4, label='Line 4', color='red')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Adjust layout to avoid overlapping titles and labels
plt.tight_layout()

# Display the plot
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a graphical representation of categorical data using rectangular bars. Each bar represents a specific category, and the length of the bar corresponds to the value or frequency of that category. Bar plots are commonly used to display and compare data between different categories.

Bar plots are used for the following purposes:

Comparison: Bar plots make it easy to visually compare the values of different categories. By looking at the length of the bars, one can quickly identify which category has a higher or lower value.

Categorical Data: Bar plots are especially useful when dealing with categorical data, where data points are divided into distinct groups or categories.

Frequency Distribution: They are used to display the frequency distribution of categorical variables, showing how often each category appears in the dataset.

Visualizing Relationships: Bar plots can also be used to show relationships between two categorical variables, like comparing the performance of different groups or conditions.

Simple Representation: Bar plots are straightforward and easy to interpret, making them accessible to a wide range of audiences.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit,color ='#7618A3',alpha =0.9)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('bar_plot')
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the five-number summary of a dataset, which includes the minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum values. The box represents the interquartile range (IQR), which is the range between Q1 and Q3, while the whiskers extend from the box to the minimum and maximum values, excluding outliers.

Box plots are used to provide a quick visual summary of the distribution of data, helping to identify the central tendency, spread, and presence of outliers in the dataset. They are especially useful when comparing multiple datasets side by side

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data, labels = ['Box 1', 'Box2'])
plt.xlabel('box')
plt.ylabel('Value')
plt.title('Box Plot')
plt.grid()
plt.show()
```
