### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot:
Use: Show the relationship between two numerical variables. Each point represents an observation in the dataset.

Line Plot:
Use: Display the trend of a numerical variable over a continuous interval or time. Useful for time-series data.

Bar Plot:
Use: Compare categories by depicting the size or frequency of a variable. Suitable for categorical data.

Box Plot:
Use: Visualize the distribution and central tendency of a numerical variable. Identify outliers and quartiles.

Violin Plot:
Use: Combines aspects of box plot and kernel density estimation. Useful for comparing the distribution of multiple variables.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
### 'pclass', y = 'age' and y = 'fare'.
### Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)

plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram
sns.histplot(data=diamonds_data, x="price", hue="cut", bins=30, multiple="stack")
plt.show()
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris datase

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot
sns.pairplot(iris_data, hue="species")
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot heatmap
sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, linewidths=.7)
plt.show()
```
